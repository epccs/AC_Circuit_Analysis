Z is unknown (V and I are known)

```python
import matplotlib.pyplot as pyplot
import numpy
import polar

tmax = 1/80
Vtoffset = 0e-3 # 10 deg: tmax*10/360 
Itoffset = (4/5)*2.5e-3 # 30 deg: tmax*30/360
freq = 1/tmax
Vpeak = 2.95*2 # Veffective*numpy.sqrt(2)
Ipeak = 2.6*0.01 # Ieffective*numpy.sqrt(2)
t = numpy.linspace(0, tmax, 100)  # 100 points over 18mSec
pyplot.figure()
pyplot.ylabel('V' + '@{:0.3g}'.format(1/tmax) + "Hz")
V = polar.Pd2R(Vpeak/numpy.sqrt(2), 360*Vtoffset/tmax)
pyplot.xlabel(polar.R2PdU(V,"V"))
pyplot.plot(t,Vpeak*numpy.sin(2*numpy.pi*freq*(t+Vtoffset)),'r')
pyplot.show()

pyplot.figure()
pyplot.ylabel('I' + '@{:0.3g}'.format(1/tmax) + "Hz")
I = polar.Pd2R(Ipeak/numpy.sqrt(2), 360*Itoffset/tmax)
pyplot.xlabel(polar.R2PdU(I,"A"))
pyplot.plot(t,Ipeak*numpy.sin(2*numpy.pi*freq*(t+Itoffset)),'b')
pyplot.show()
Z = V/I
print("Impedance:" + polar.R2PdU(Z,"-Ohm"))
if(numpy.imag(Z)>0):
    print("L: {:0.3g}".format(numpy.imag(Z)/(numpy.pi*2*1/tmax)) + "H")
if(numpy.imag(Z)<0):
    print("C: {:0.3g}".format(-1/(numpy.imag(Z)*numpy.pi*2*(1/tmax)) ) + "F")
if(numpy.real(Z)>0):
    print("R: {:0.3g}".format(numpy.real(Z)) + "-Ohm")
```

I is unknown (V and Z are know)

```python
tmax = 0.014 
Vtoffset = -2e-3 # 10 deg: tmax*10/360 
freq = 1/tmax
Vpeak = 15.2 # Veffective*numpy.sqrt(2)
t = numpy.linspace(0, tmax, 100)  # 100 points over 18mSec
pyplot.figure()
pyplot.ylabel('V' + '@{:0.3g}'.format(1/tmax) + "Hz")
V = polar.Pd2R(Vpeak/numpy.sqrt(2), 360*Vtoffset/tmax)
pyplot.xlabel(polar.R2PdU(V,"V"))
pyplot.plot(t,Vpeak*numpy.sin(2*numpy.pi*freq*(t+Vtoffset)),'r')
pyplot.show()

Z = polar.Pd2R(200,0) # 200 Ohm

pyplot.figure()
pyplot.ylabel('I' + '@{:0.3g}'.format(1/tmax) + "Hz")
I = V/Z
pyplot.xlabel(polar.R2PdU(I,"A"))
Itoffset = tmax * (numpy.rad2deg(numpy.angle(I))/360)
pyplot.plot(t,numpy.abs(I)*numpy.sqrt(2)*numpy.sin(2*numpy.pi*freq*(t+Itoffset)),'b')
pyplot.show()
Z = V/I
print("Impedance:" + polar.R2PdU(Z,"-Ohm"))
if(numpy.imag(Z)>0):
    print("L: {:0.3g}".format(numpy.imag(Z)/(numpy.pi*2*1/tmax)) + "H")
if(numpy.imag(Z)<0):
    print("C: {:0.3g}".format(-1/(numpy.imag(Z)*numpy.pi*2*(1/tmax)) ) + "F")
if(numpy.real(Z)>0):
    print("R: {:0.3g}".format(numpy.real(Z)) + "-Ohm")
```

```python
tmax = 0.020 
Itoffset = 4e-3 # 10 deg: tmax*10/360 
freq = 1/tmax
Ipeak = 35e-3 # Ieffective*numpy.sqrt(2)
t = numpy.linspace(0, tmax, 100)  # 100 points over 18mSec
pyplot.figure()
pyplot.ylabel('I' + '@{:0.3g}'.format(1/tmax) + "Hz")
I = polar.Pd2R(Ipeak/numpy.sqrt(2), 360*Itoffset/tmax)
pyplot.xlabel(polar.R2PdU(I,"A"))
pyplot.plot(t,numpy.abs(I)*numpy.sqrt(2)*numpy.sin(2*numpy.pi*freq*(t+Itoffset)),'b')
pyplot.show()

Z = polar.Pd2R(1/(2*numpy.pi*(1/tmax)*15e-6),-90) # capacitor 1/(2*pi*freq*15uF)

pyplot.figure()
pyplot.ylabel('V' + '@{:0.3g}'.format(1/tmax) + "Hz")
V = I*Z
pyplot.xlabel(polar.R2PdU(V,"V"))
Vtoffset = tmax * (numpy.rad2deg(numpy.angle(V))/360)
pyplot.plot(t,numpy.abs(V)*numpy.sqrt(2)*numpy.sin(2*numpy.pi*freq*(t+Vtoffset)),'r')
pyplot.show()
Z = V/I
print("Impedance:" + polar.R2PdU(Z,"-Ohm"))
if(numpy.imag(Z)>0):
    print("L: {:0.3g}".format(numpy.imag(Z)/(numpy.pi*2*1/tmax)) + "H")
if(numpy.imag(Z)<0):
    print("C: {:0.3g}".format(-1/(numpy.imag(Z)*numpy.pi*2*(1/tmax)) ) + "F")
if(numpy.real(Z)>0):
    print("R: {:0.3g}".format(numpy.real(Z)) + "-Ohm")
```
